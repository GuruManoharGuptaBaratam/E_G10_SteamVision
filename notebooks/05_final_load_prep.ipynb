05 Final Data Load Preparation - The Success Predictor

Problem Statement: What factors drive commercial success, and how can we predict it?

In this final notebook, we build a Success Metric based on the findings from previous analyses. This number helps developers understand their market performance relative to the competition.

1. Setup and Loading Data

```python
import pandas as pd

df = pd.read_csv('../data/processed/cleaned_game_data.csv')
df['release'] = pd.to_datetime(df['release'])

print(f'Finalizing analysis for {len(df):,} games.')
```

--- 
2. Ranking Success (The Success Score)

Goal: Create a single number (0-100) that shows total popularity.

We combine all_time_peak (usage) and total_reviews (reach) into a Commercial Success Index (CSI).

```python
p_norm = (df['all_time_peak'] / df['all_time_peak'].max())
r_norm = (df['total_reviews'] / df['total_reviews'].max())

df['success_score'] = (p_norm * 0.4 + r_norm * 0.6) * 100
df['success_score'] = df['success_score'].round(2)

print('Commercial Success Index (CSI) computed.')
```

Insight: Why this metric? 
It merges two different success signs into one easy-to-read score. A developer can now compare their score against the genre average.

--- 
3. Market Outcomes (Hit vs. Fail)

Goal: Define what failure looks like to better predict it.

```python
def classify(row):
    if row['all_time_peak'] >= 500 or row['total_reviews'] >= 250: return 'Hit'
    if row['all_time_peak'] >= 100 or row['total_reviews'] >= 50: return 'Average'
    return 'Niche/Fail'

df['market_outcome'] = df.apply(classify, axis=1)
print('Market outcomes assigned.')
print(df['market_outcome'].value_counts())
```

Insight: Why this classification? 
Most Steam games reach very few people. Knowing the Danger Zone helps publishers assess risk before funding a project.

--- 
4. Final Dataset Export

Goal: Keep only columns that help answer our problem statement.

```python
final_cols = [
    'appid', 'game', 'release', 'primary_genre', 'developer', 
    'publisher', 'rating', 'total_reviews', 'all_time_peak', 
    'success_score', 'market_outcome'
]

df[final_cols].to_csv('../data/processed/final_success_analysis.csv', index=False)
print("Final Success Analysis File Saved to 'data/processed/'.")
```

5. Project Conclusion

The analysis proves that while high quality is a requirement, mass success is driven by community buzz and genre positioning. The CSI and Outcomes tools now provide a clear path for predicting market potential.
